# Automating Crypto API Scraping
Data scraped from: https://coinmarketcap.com/api/

```python
#importing libraries from API documentation
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest' #change domain based on API documentation
parameters = {
  'start':'1',
  'limit':'15', # I changed this to 15, otherwise the file is going to be huge
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': 'b978fa69-a90e-4f2e-83f7-96b4b7f36e6c', #change api key to your own registered api key
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
#importing pandas to create dataframe
import pandas as pd
df = pd.json_normalize(data['data'])
df
```

# Creating a function to scrape data and export to csv

```python
def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'b978fa69-a90e-4f2e-83f7-96b4b7f36e6c',
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)

    df2 = pd.json_normalize(data['data'])
    df2['timestamp'] = pd.to_datetime('now')
    df_new_pull = pd.DataFrame(df2)
    df = pd.concat([df,df_new_pull])
    if not os.path.isfile('/Users/joeytan/Documents/Data Analyst/Portfolio/PortfolioProject4_Crypto/crypto_data.csv'):
        df.to_csv('/Users/joeytan/Documents/Data Analyst/Portfolio/PortfolioProject4_Crypto/crypto_data.csv', header = 'column_names')
    else:
        df.to_csv('/Users/joeytan/Documents/Data Analyst/Portfolio/PortfolioProject4_Crypto/crypto_data.csv', mode = 'a', header = False)
```

# Automating function to pull data every 60s

```python
import os
from time import time
from time import sleep

for i in range(333):
    api_runner()
    print('API Runner Completed')
    sleep(60) #sleep for 1 min
exit()
```
